```python
import matplotlib.pyplot as plt
import numpy as np
from PIL import Image
```

```python
path_to_dataset = "/home/yangfei/clean-pvnet/data/custom/"
```

```python
# Camera.txt: Intrinsic parameters
with open(path_to_dataset+"/camera.txt", "r") as file:
    camera_txt = file.read()
print(camera_txt)
```

```python
# diameter.txt : Diameter of object(Unit: meter)
with open(path_to_dataset+"/diameter.txt", "r") as file:
    diameter_txt = file.read()
print(diameter_txt)
```

```python
# mask
mask_path = path_to_dataset + "mask/0.png"
mask = Image.open(mask_path)
print("Mask file path:", "/".join(mask_path.split('/')[-3:]))
print("Mask image format:", mask.format)
print("Mask image size:", mask.size)
print("Mask image mode:", mask.mode)

plt.imshow(mask)
plt.axis('off')
plt.show()
```

```python
# image
img_path = path_to_dataset + "rgb/0.jpg"
img = Image.open(img_path)
print("Image file path:", "/".join(img_path.split('/')[-3:]))
print("Image format:", img.format)
print("Image size:", img.size)
print("Image mode:", img.mode)

plt.imshow(img)
plt.axis('off')
plt.show()
```

```python
pose_path = path_to_dataset + "pose/pose0.npy"
pose = np.load(pose_path)

print("Pose file path:", "/".join(pose_path.split('/')[-3:]))
print("Pose annotation is: ")
print(pose)
```
